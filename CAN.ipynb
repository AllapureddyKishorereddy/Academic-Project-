```python
import pandas as pd

# File paths for the 4 CSV files you want to combine
file_paths = [
    r'Attack_free_with_target.csv',
    r'DoS_attack_with_target.csv',
    r'Fuzzy_attack_with_target.csv',
    r'Impersonation_attack_with_target.csv'
]

# Initialize an empty list to hold the DataFrames
dfs = []

# Loop through each file and read it into a DataFrame
for file_path in file_paths:
    df = pd.read_csv(file_path)
    dfs.append(df)

# Concatenate all DataFrames into one
combined_df = pd.concat(dfs, ignore_index=True)

# Save the combined DataFrame to a new CSV file
combined_df.to_csv('CNN_Intrusion.csv', index=False)

# Display the first few rows of the combined dataset to confirm
combined_df.head()
```
